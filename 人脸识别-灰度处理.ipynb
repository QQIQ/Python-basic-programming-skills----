```python
#img=cv2.open("F:\image\20160125163004_KeSGf.jpeg")
#img=Image.open("F:/image/20160125163004_KeSGf.jpeg")
img=Image.open("F:/mena_suvari-027.jpg")
img=np.array(img)
plt.imshow(img)
plt.show()
```

```python
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_cascade.load('D:/pythonana/envs/py3/Lib/site-packages/cv2/data/haarcascade_frontalface_alt_tree.xml')
img=cv2.imread('F:/Img8940498_n.jpg')
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
faces=face_cascade.detectMultiScale(gray,1.3,5)
print(faces)
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)

#cv2.imshow('image',img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
for(x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),5)
    roi_gray=gray[y:y+h,x:x+w]
    roi_color=img[y:y+h,x:x+2]
#img=np.array(img)
plt.imshow(img)
plt.show()
```
